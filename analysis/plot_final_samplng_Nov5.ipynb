```python
import pandas as pd

data = pd.read_parquet("./final_sampling_results.parquet.gzip")
data = data.rename(columns={"sampler": "Sampling scheme", "rtt": "RTT"})
data
```

```python
mean_per_step = (
    data
    .groupby(["ground_truth", "Sampling scheme", "neuroticism", "RTT", "repetition"], observed=True)
    .mean()
    .reset_index()
)

mean_per_step.groupby(["ground_truth", "Sampling scheme", "neuroticism", "RTT",], observed=True).mean()
```

```python
import numpy as np

def select_data_by_ground_truth(data: pd.DataFrame, ground_truth: str) -> pd.DataFrame:
    data = data[
        (data["ground_truth"] == ground_truth)
        & np.isin(data["neuroticism"], (ground_truth, "n/a"))
    ].copy()

    data["ground_truth"] = data["ground_truth"].cat.remove_unused_categories()
    data["neuroticism"] = data["neuroticism"].cat.remove_unused_categories()

    # print(ground_truth, data["ground_truth"].unique(), data["neuroticism"].unique())

    return data
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.ticker as tkr
import pandas as pd
import seaborn as sns

sns.set_style("whitegrid")
sns.set_context("paper", font_scale=2)
sns.set_palette("Dark2")


# scheme color and marker mappings
marker_mapping = {
    "Energy-optimized\naperiodic": "p",
    "Greedy": "^",
    "Offline optimum": "v",
    "Sample-count-optimized\naperiodic": "s",
    "Wang et. al. 2019\nGaussian fit": "o"
}

with sns.color_palette() as palette:
    color_mapping = {
        "Energy-optimized\naperiodic": palette[4],
        "Greedy": palette[1],
        "Offline optimum": palette[0],
        "Sample-count-optimized\naperiodic": palette[3],
        "Wang et. al. 2019\nGaussian fit": palette[2]
    }

marker_mapping, color_mapping
```

```python
def plot_mean_something(data: pd.DataFrame, column: str, *args, **kwargs):
    ax = plt.gca()
    gts = data["ground_truth"].unique()
    assert gts.size == 1

    ground_truth = gts[0]
    selected_data = select_data_by_ground_truth(data, ground_truth)

    sns.lineplot(
        data=selected_data,
        errorbar=("ci", 95),
        err_style="bars",
        x="RTT",
        y=column,
        hue="Sampling scheme",
        style="Sampling scheme",
        markers=marker_mapping,
        markersize=10,
        ax=ax,
        palette=color_mapping
    )

mean_without_energy = mean_per_step[mean_per_step["Sampling scheme"] != "Energy-optimized\naperiodic"].copy()
mean_without_energy["Sampling scheme"] = mean_without_energy["Sampling scheme"].cat.remove_unused_categories()

fg = sns.FacetGrid(mean_without_energy, col="ground_truth", aspect=1.5, height=5)
fg.map_dataframe(plot_mean_something, column="num_samples")
fg.set_axis_labels("Round-Trip Time", "Mean number of samples\ncaptured per step")
fg.set(yticks=[2**i for i in range(5)])
fg.add_legend(title="Sampling scheme")
for legend_handle in fg.legend.legendHandles:
    legend_handle.set_markersize(10)

plt.yscale("log", base=2)
for ax in fg.axes.flat:
    ax.xaxis.set_major_formatter(tkr.FuncFormatter(lambda x, p: f"{x:0.1f}s"))
    ax.yaxis.set_major_formatter(tkr.FuncFormatter(lambda y, p: f"{y:0.0f}"))

fg.set_titles(col_template="Timing model & estimator neuroticism: {col_name}")
plt.show()
```

```python
mean_without_samples = mean_per_step[mean_per_step["Sampling scheme"] != "Sample-count-optimized\naperiodic"].copy()
mean_without_samples["Sampling scheme"] = mean_without_samples["Sampling scheme"].cat.remove_unused_categories()

fg = sns.FacetGrid(mean_without_samples, col="ground_truth", aspect=1.5, height=5)
fg.map_dataframe(plot_mean_something, column="energy")
fg.set_axis_labels("Round-Trip Time", "Mean per step\nenergy consumption")
# fg.set(yticks=np.linspace(0.0, 0.7, 8))
fg.set(ylim=(None, 0.68))
fg.add_legend(title="Sampling scheme")
for legend_handle in fg.legend.legendHandles:
    legend_handle.set_markersize(10)

for ax in fg.axes.flat:
    ax.xaxis.set_major_formatter(tkr.FuncFormatter(lambda x, p: f"{x:0.1f}s"))
    ax.yaxis.set_major_formatter(tkr.FuncFormatter(lambda y, p: f"{y * 1000:1.0f} mJ"))

fg.set_titles(col_template="Timing model & estimator neuroticism: {col_name}")

plt.show()
```

```python
from collections import deque
from typing import Sequence
import pingouin as pg

ix = pd.IndexSlice

def format_rtt(v: float) -> str:
    return f"{v:.1f}s"

def plot_diff_wrt_all_others(data: pd.DataFrame, column: str, scheme: str, ref_schemes: str, *args, **kwargs):
    gts = data["ground_truth"].unique()
    assert gts.size == 1
    ground_truth = gts[0]
    selected_data = select_data_by_ground_truth(data, ground_truth)
    selected_data = selected_data[
        (selected_data["Sampling scheme"] == scheme)
        | (np.isin(selected_data["Sampling scheme"], ref_schemes))
    ].copy()

    result_rows = deque()
    for rtt, rtt_df in selected_data.groupby("RTT"):
        ref_values = rtt_df.loc[rtt_df["Sampling scheme"] == scheme, column].to_numpy()
        ref_mean = ref_values.mean()
        assert ref_values.size > 0

        for (comp_scheme, neuro), comp_df in rtt_df[np.isin(rtt_df["Sampling scheme"], ref_schemes)].groupby(["Sampling scheme", "neuroticism"], observed=True):
            samples = comp_df[column].to_numpy()
            assert samples.size > 0

            smean = samples.mean()
            ttest = pg.ttest(ref_values, samples, alternative="two-sided")

            diff_raw = ref_mean - smean
            diff = diff_raw / smean
            ci = ttest.at["T-test", "CI95%"]

            pval = ttest.at["T-test", "p-val"]
            sig_pval = pval < 0.05

            result_rows.append({
                "rtt": rtt,
                "Sampling scheme": comp_scheme,
                "neuroticism": neuro,
                "diff": diff,
                "diff_raw": diff_raw,
                "p < 0.05" : sig_pval,
                "CI95_Low" : ci[0] / smean,
                "CI95_High": ci[1] / smean,
            })

    results = pd.DataFrame(result_rows)




    # selected_data = selected_data[np.isin(selected_data["RTT"], rtts)].copy()

    # diff_df = selected_data.groupby(["RTT"]).apply(diff_wrt_ref).unstack()
    # diff_df.columns = diff_df.columns.droplevel(0)
    results = (
        results
        # .reset_index()
        .set_index(["Sampling scheme", "neuroticism", "rtt"], verify_integrity=True)
        [["CI95_Low", "CI95_High", "diff"]]
        .stack()
        .reset_index()
        .rename(columns={"level_3": "metric", 0: "value"})
    )

    sns.lineplot(
        estimator=lambda e: sorted(e)[1],
        errorbar=lambda e: (min(e), max(e)),
        err_style="bars",
        ax=plt.gca(),
        data=results,
        x="rtt",
        y="value",
        hue="Sampling scheme",
        style="Sampling scheme",
        markers=marker_mapping,
        markersize=10,
        palette=color_mapping
    )

    # sns.barplot(
    #     estimator=lambda e: sorted(e)[1],
    #     errorbar=lambda e: (min(e), max(e)),
    #     ax=plt.gca(),
    #     data=diff_df,
    #     x="value",
    #     y="RTT",
    #     hue="Sampling scheme",
    # )

    # return diff_df

fg = sns.FacetGrid(mean_without_energy, col="ground_truth", aspect=1.5, height=5)
fg.map_dataframe(
    plot_diff_wrt_all_others,
    column="num_samples",
    scheme="Sample-count-optimized\naperiodic",
    ref_schemes=["Greedy", "Offline optimum", "Wang et. al. 2019\nGaussian fit"]
)
fg.set_axis_labels("Round-Trip Time", "Difference in number of\ncaptured samples per step")
fg.add_legend(title="Reference scheme")
for legend_handle in fg.legend.legendHandles:
    legend_handle.set_markersize(10)

# xticks = np.array([2, 4, 6, 8, 10, 12, 14, 16, 18]) - 1
# fg.set(xticks=np.linspace(0, 16, 5))
# plt.xscale("log", base=2)
for ax in fg.axes.flat:
    ax.yaxis.set_major_formatter(tkr.FuncFormatter(lambda v, p: f"{v:+0.0%}" if v > 0 else f"{v:0.0%}"))
    ax.xaxis.set_major_formatter(tkr.FuncFormatter(lambda v, p: f"{v:0.1f}s"))


fg.set_titles(col_template="Timing model & estimator neuroticism: {col_name}")
plt.show()
```
