```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as tkr
import numpy as np
import pandas as pd
from generate_verification_data import models

sns.set_style("whitegrid")
sns.set_context("paper", font_scale=2)
sns.set_palette("Dark2")
```

```python
ttf_results = pd.read_parquet("verification/ttf.parquet")
ttf_results
```

```python
duration_results = pd.read_parquet("verification/duration.parquet")
duration_results
```

```python
transition_results = pd.read_parquet("verification/transitions.parquet")
transition_results["Transition"] = transition_results["Transition"].replace({"Same level": "No change"})
transition_results = transition_results.rename(columns={"Transition": "Change in TTF"})
transition_results
```

```python
fg = sns.catplot(
    kind="bar",
    # join=False,
    dodge=True,
    data=ttf_results,
    y="Model",
    order=models.keys(),
    x="Execution time",
    hue="Time-to-Feedback",
    hue_order=["low", "medium", "high"],
    aspect=1.5,
)
# fg.set(xlim=(None, 8))
# fg.set_axis_labels("Execution time [s]", "Configuration")
# fg.legend.set_title("Time-to-feedback")
fg.set_ylabels("Configuration")
for ax in fg.axes.flat:
    ax.xaxis.set_major_formatter(tkr.FuncFormatter(lambda y, p: f"{y:2.1f}s"))
plt.show()
```

```python
ix = pd.IndexSlice

neuro_ttfs = (
    ttf_results
    .groupby(["Model", "Neuroticism", "Time-to-Feedback"])["Execution time"]
    .mean()
    .groupby(["Neuroticism", "Time-to-Feedback"])
    .mean()
)
(neuro_ttfs.loc[ix[:, "high"]] - neuro_ttfs.loc[ix[:, "low"]]) / neuro_ttfs.loc[ix[:, "low"]]
```

```python
fg = sns.relplot(
    kind="line",
    err_style="band",
    errorbar=("ci", 95),
    data=duration_results,
    x="Step",
    y="Execution time",
    hue="Configuration",
    hue_order=models.keys(),
    style="Time-to-Feedback",
    style_order=["low", "high"],
    markers=["v", "^"],
    markersize=15,
    aspect=0.5,
    height=10,
)
for ax in fg.axes.flat:
    ax.xaxis.set_major_locator(tkr.MaxNLocator(integer=True))
    ax.yaxis.set_major_formatter(tkr.FuncFormatter(lambda y, p: f"{y:0.1f}s"))
fg.set_axis_labels("Step", "Execution time")
# fg.set(ylim=(4, 8))

for legend_handle in fg.legend.legendHandles:
    legend_handle.set_markersize(10)
    # print(dir(legend_handle))
    # break

plt.show()
```

```python
duration_results["Neuroticism"] = "High"
duration_results.loc[duration_results.Configuration.str.contains("[Ll]ow neuro"), "Neuroticism"] = "Low"

ttfs = sorted(duration_results["TTF (Raw)"].unique())
def label_ttf_s(ttf: float) -> str:
    return f"{ttf}s"

duration_results["TTF"] = duration_results["TTF (Raw)"].apply(label_ttf_s)
duration_results["TTF"] = duration_results["TTF"].astype(
    pd.CategoricalDtype(map(label_ttf_s, ttfs), ordered=True)
)
duration_results
```

```python
from matplotlib.ticker import MaxNLocator

fg = sns.relplot(
    kind="line",
    err_style="band",
    errorbar=("ci", 95),
    data=duration_results,
    x="Step",
    y="Execution time",
    col="Neuroticism",
    # hue_order=models.keys(),
    hue="Time-to-Feedback",
    hue_order=["low", "medium", "high"],
    # dodge=True,
    # markers=["v", "^"],
    # markersize=15,
    # aspect=0.5,
    # height=10,
)
for ax in fg.axes.flat:
    ax.xaxis.set_major_locator(tkr.MaxNLocator(integer=True))
#     ax.yaxis.set_major_formatter(tkr.FuncFormatter(lambda y, p: f"{y:0.1f}s"))
fg.set_axis_labels("Step", "Execution Time [s]")
fg.set_titles(col_template="{col_name} {col_var}")
# fg.set(ylim=(4, 8))

# for legend_handle in fg.legend.legendHandles:
#     legend_handle.set_markersize(10)
    # print(dir(legend_handle))
    # break

plt.show()
```
